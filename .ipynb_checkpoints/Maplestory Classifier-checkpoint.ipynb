```python
def image_to_array(folder_path):
    files = os.listdir(folder_path)
    images = [file for file in files if file.endswith('.JPG') or file.endswith('.png')]
    image_list = [] 
    for image in images:
        img = cv2.imread(os.path.join(folder_path, image))
        image_list.append(img)
    return image_list
```
